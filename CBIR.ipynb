# Spirit Label CBIR - BiBirra Clone

```python
# import necessary packages
import urllib
import os
import time
import shutil
import numpy as np
import cv2
import pandas as pd
```

```python
orb = cv2.ORB_create(nfeatures=1500)

img_spirit = [file for file in os.listdir('referenceImages/') if file.endswith('.png')]
# img_spirit = np.random.choice(img_spirit, len(img_spirit))

df = pd.DataFrame()
for image in img_spirit:
    img = cv2.imread("referenceImages/"+image, cv2.IMREAD_UNCHANGED)
    keypoints, desc = orb.detectAndCompute(img, None)
    df1 = pd.DataFrame({image: {"keypoints" : keypoints, "descriptor": desc }})
    df = pd.concat([df, df1], axis=1)
```

```python
spirits_db = df.T
spirits_db
```

```python
def query(image):
    img = cv2.imread(image, cv2.IMREAD_UNCHANGED)
    keypoints, desc = orb.detectAndCompute(img, None)
    bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
    spirits_db[f"matches_{image}"] = spirits_db["descriptor"].apply(lambda x: bf.match(desc, x))
    spirits_db[f"closest_{image}"] = spirits_db[f"matches_{image}"].apply(get_distance_sum)
    closest = spirits_db[spirits_db[f"closest_{image}"] == spirits_db[f"closest_{image}"].min()]
    return closest.index[0]

def get_distance_sum(match):
    dist = 0
    match = sorted(match, key=lambda x: x.distance)
    for m in match[:100]:
        dist += m.distance
    return dist

for image in os.listdir("queryImages/"):
    if image.endswith(".jpg") or image.endswith(".png"):
        img = cv2.imread(f"queryImages/{image}", cv2.IMREAD_GRAYSCALE)
        print(image, "------", query(f"queryImages/{image}"))
```
